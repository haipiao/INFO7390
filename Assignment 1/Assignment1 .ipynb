```python
import random

def birthday(peopleNum):
    bingo = 0
    bingoIndex = 100 # 100 means the date of Oct. 3rd
    list = []
    n = 0
    while n < peopleNum:
        list.append(random.randint(1,365))
        n = n + 1
    
    bingoNum = 0
    for l in list:
        if(l == bingoIndex):
            bingoNum = bingoNum + 1
    
    if(bingoNum > 1):
        return True
    else:
        return False
```

```python
def testBirthday(peopleNum, times):
    p_total = 0
    bingoTimes = 0
    i = 0
    while i < times:
        if(birthday(peopleNum)== True):
            bingoTimes = bingoTimes + 1
        i = i + 1
    p_total = bingoTimes / times
    return p_total
```

##### Q8 (5 Points)   Write python code to simulate question 7.

```python
import random

def vote(n):
    M = n*0.3 #1
    I = n*0.2 #2
    B = n*0.5 #3
    lazy = n* 0.25
    
    lazy_M = 0
    lazy_I = 0
    lazy_B = 0
    
    i = 0
    while i<lazy:
        getNum = random.randint(1,3)
        if(getNum == 1):
            lazy_M = lazy_M + 1
        if(getNum == 2):
            lazy_I = lazy_I + 1
        if(getNum == 3):
            lazy_B = lazy_B + 1
        i = i + 1
        
        
    return [(M-lazy_M)/(n-lazy),(I-lazy_I)/(n-lazy),(B-lazy_B)/(n-lazy)]
```

```python
times = 10000
p_M = 0
p_I = 0
p_B = 0
i = 0

while i <times:
    list = vote(1000)
    p_M = p_M + list.pop()
    p_I = p_I + list.pop()
    p_B = p_B + list.pop()
    i = i+1

p_M = p_M / times
p_I = p_I / times
p_B = p_B / times

print('Effective vote rate')
print('M=',p_M)
print('I=',p_I)
print('B=',p_B)
```

```python
import random

def door(n,k):
    bingo = 0
    bingoIndex = random.randint(0,n-k-1)
    list = []
    #print(bingoIndex)
    i = 0
    while i < n:
        if (i == bingoIndex):
            list.append(1)
        else:
            list.append(0)
        i = i + 1
    
    #print(list)
    chooseIndex = random.randint(0,n-k-1)
    
    if(chooseIndex == bingoIndex):
        return True
    else:
        return False
```

```python
def testDoor(n,k,times):
    p_total = 0
    bingoTimes = 0
    i = 0
    while i < times:
        if (door(n,k)== True):
            bingoTimes = bingoTimes + 1
        i = i + 1
    p_total = bingoTimes / times 
    return p_total
```
